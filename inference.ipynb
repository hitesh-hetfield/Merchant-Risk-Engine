```python
import joblib

churn_model = joblib.load("artifacts/churn_model.pkl")
features = joblib.load("artifacts/feature_names.pkl")

print(features)
```

```python
merchant_input = {
    "TxFreq": 3,
    "TotalTx90": 1200.0,
    "AvgTx90": 400.0,
    "Recency": 12
}
```

```python
import pandas as pd 

input_df = pd.DataFrame([merchant_input])
input_df = input_df[features]

churn_prob = churn_model.predict_proba(input_df)[0][1]
churn_flag = int(churn_prob >= 0.5)
```

```python
print(churn_prob, churn_flag)
```

```python
# Defining Risk Buckets

LOW_RISK = 0.30
HIGH_RISK = 0.60

def assign_risk_bucket(churn_prob):
    if churn_prob < LOW_RISK:
        return "LOW"
    elif churn_prob < HIGH_RISK:
        return "MEDIUM"
    else:
        return "HIGH"
    
def recommended_action(risk_bucket):
    if risk_bucket == "LOW":
        return "NAR"
    elif risk_bucket == "MEDIUM":
        return "Monitor Usage"
    else:
        return "Trigger Retention Offer"
```

```python
risk_bucket = assign_risk_bucket(churn_prob)
action = recommended_action(risk_bucket)

print(f"Risk: {risk_bucket}, Action: {action}")
```

```python
result = {
    "churn_probability": round(churn_prob, 3),
    "churn_flag": churn_flag,
    "risk_bucket": risk_bucket,
    "recommended_action": action
}

print(result)
```
